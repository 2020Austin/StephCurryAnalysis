## Data Cleaning: convert to numerical values

```python
import pandas as pd
import numpy as np
```

```python
# Read csv and get info

filename = "stephcurry.csv"

df = pd.read_csv(filename)

df.info()
```

```python
df1 = df.drop(df.columns[[0, 1, 2, 17, 19, 20]], axis=1) # Drop Dates, Opponent, Score, Type (Regular Season vs Conference), Score GS, Score Opponent

df1 = df1.dropna() #Drop na
df1
```

```python
# Convert L to 0, W to 1
    
df1['Result'].replace({ 'L' : 0, "W" : 1}, inplace=True)
df1
```

```python
# Tidy!
df1.to_csv("cleaned_stephcurry.csv")
```
